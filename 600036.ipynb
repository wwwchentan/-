```python
import tushare as ts
data = ts.get_hist_data("600036")
```

```python
data
```

```python
X_train = []
y_train = []
for i in range(10, data.shape[0]):
    tmp = []
    tmp.append(data.ix[i]['open'])
    tmp.append(data.ix[i]['high'])
    tmp.append(data.ix[i]['close'])
    tmp.append(data.ix[i]['low'])
    X_train.append(tmp)
    tmp = (data.ix[i]['close'] - data.ix[i]['open']) / data.ix[i]['open']
    y_train.append(tmp)
    
X_test = X_train[:10]
    
# print(X_train.shape)
# print(y_train.shape)
```

```python
from sklearn.linear_model import  LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
print(y_pred)
```
